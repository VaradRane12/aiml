```python
import numpy as np

# --- Cell 1: Sample Data ---
X_train = np.array([2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1])
y_train = np.array([21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17])

# --- Cell 2: Model Training (Fit Logic) ---
b0 = 0
b1 = 0

x_mean = np.mean(X_train)
y_mean = np.mean(y_train)

numerator = np.sum((X_train - x_mean) * (y_train - y_mean))
denominator = np.sum((X_train - x_mean)**2)

b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)

print(f"Model Parameters:")
print(f"Intercept (b0): {b0}")
print(f"Slope (b1): {b1}")

# --- Cell 3: Prediction (Predict Logic) ---
y_pred = b0 + b1 * X_train
print(f"\nPredictions on training data: {y_pred}")

# --- Cell 4: Model Evaluation ---
# Mean Squared Error
mse = np.mean((y_train - y_pred)**2)

# R2 Score
y_mean_r2 = np.mean(y_train)
ss_total = np.sum((y_train - y_mean_r2)**2)
ss_residual = np.sum((y_train - y_pred)**2)
r2 = 1 - (ss_residual / ss_total)

print(f"\nModel Evaluation:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) Score: {r2}")

# --- Cell 5: Example Prediction ---
study_hours_new = 6.0
predicted_score = b0 + b1 * study_hours_new
print(f"\nPrediction for {study_hours_new} hours: {predicted_score}")

```
