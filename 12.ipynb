```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("studs.csv")
```

```python
df.head()
```

```python
x = df["hours_studied"]
y = df["exam_score"]
```

```python
n = len(x)
```

```python
x_mean = x.mean()
y_mean = y.mean()
```

```python
num = np.sum((x-x_mean) * (y-y_mean))
den = np.sum((x-x_mean)**2)
b1 = num/den
b0 = y_mean -b1*x_mean
```

```python
y_pred = b0+b1*x
```

```python
print(y_pred)
```

```python
mse = np.sum((y-y_pred)**2)
```
