```python
from collections import deque

def bfs_shortest_path(maze, start, goal):
    rows = len(maze)
    cols = len(maze[0])

    directions = [(0,1), (0,-1), (1,0), (-1,0)]

    queue = deque([start])
    visited = set([start])
    
    parent = {start: None}

    print("Start BFS")
    print("Initial Queue:", queue)
    print("Initial Visited:", visited)
    print()

    while queue:
        x, y = queue.popleft()
        print(f"Popped from queue: {(x, y)}")

        if (x, y) == goal:
            print("Goal reached. Reconstructing path...")
            path = []
            curr = goal
            while curr:
                path.append(curr)
                curr = parent[curr]
            path.reverse()
            return path

        for dx, dy in directions:
            nx, ny = x + dx, y + dy

            print(f" Checking neighbor: {(nx, ny)}")

            if 0 <= nx < rows and 0 <= ny < cols:
                print("  Inside bounds")
                if maze[nx][ny] == 0 and (nx, ny) not in visited:
                    print("  Free cell and not visited. Adding to queue")
                    visited.add((nx, ny))
                    parent[(nx, ny)] = (x, y)
                    queue.append((nx, ny))
                    print("  Queue now:", queue)
                    print("  Visited now:", visited)
                else:
                    print("  Wall or already visited")
            else:
                print("  Out of bounds")
        print()

    return None


maze = [
    [0, 0, 1, 0],
    [1, 0, 1, 0],
    [0, 0, 0, 0], 
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 3)

path = bfs_shortest_path(maze, start, goal)
print("Shortest Path:", path)
```
